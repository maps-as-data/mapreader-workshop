# Download maps

```python
from mapreader import SheetDownloader
```

```python
downloader = SheetDownloader(
    metadata_path="./sixinch.json",
    download_url = "https://geo.nls.uk/mapdata3/os/6inchfirst/{z}/{x}/{y}.png",
    )
```

```python
# extract publication dates from the metadata
downloader.extract_published_dates(["properties", "YEAR"])
```

```python
# this is a preselected list of maps in/around Glasgow to use in this workshop
ids = [
    "74427695",
    "74427696",
    "74427699",
    "74427700",
    "74427706",
    "74427705",
]

for id in ids:
	downloader.query_map_sheets_by_string(id, append=True)
```

```python
downloader.print_found_queries()
```

```python
downloader.plot_queries_on_map(map_extent="uk", add_id=False)
```

```python
# set zoom level and download the maps
downloader.get_grid_bb(16)
downloader.download_map_sheets_by_queries(force=True)
```
